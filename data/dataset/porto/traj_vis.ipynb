```python
import pickle as pk
import random
import folium 
import numpy as np
```

```python
trajs = pk.load(open('porto_coord_10000.pkl', 'rb'))
grid_trajs = pk.load(open('porto_grid_10000.pkl', 'rb'))
```

### Calculate the Statistical Data 

```python
lons = []
lats = []
for traj in trajs:
    for point in traj:
        lons.append(point[0])
        lats.append(point[1])
lats = np.array(lats)
lons = np.array(lons)
gps_stat = { "mean_lat": lats.mean(), "mean_lon": lons.mean(), "std_lat": lats.std(), "std_lon": lons.std()}
print(gps_stat)
```

```python
len_stat = [len(traj) for traj in trajs]
print(f"Traj: max len: {max(len_stat)}, min len: {min(len_stat)}, mean len: {np.mean(len_stat)}")
len_stat = [len(traj) for traj in grid_trajs]
print(f"GridTraj: max len: {max(len_stat)}, min len: {min(len_stat)}, mean len: {np.mean(len_stat)}")
```

### Check the correctness of the distance matrix.

```python
import traj_dist.distance as tdist
hausdorff_matrix = pk.load(open('dis_matrix_haus_10k.pkl', 'rb'))
dtw_matrix = pk.load(open('dis_matrix_dtw_10k.pkl', 'rb'))
fret_matrix = pk.load(open('dis_matrix_dfrec_10k.pkl', 'rb'))
print(hausdorff_matrix.shape, dtw_matrix.shape, fret_matrix.shape)
```

```python
i, j = random.sample(range(10000), 2)
print("hausdorff")
print(hausdorff_matrix[i][j])
print(tdist.hausdorff(np.array(trajs[i]), np.array(trajs[j])))
print("dtw")
print(dtw_matrix[i][j])
print(tdist.dtw(np.array(trajs[i]), np.array(trajs[j])))
print("frechet")
print(fret_matrix[i][j])
print(tdist.discret_frechet(np.array(trajs[i]), np.array(trajs[j])))
```

### Trajectory visualization 

```python
# random sample 100 trajs to plot 
folium_map = folium.Map(location=[41.1647,-8.6203], tiles='CartoDB dark_matter')
for traj in random.sample(trajs, 1000):
    traj = [[y, x] for (x, y) in traj]
    folium.PolyLine(traj,color='orange', weight=1).add_to(folium_map)
folium_map
```
